```python
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
```

```python
corrente_eletrica = np.array([5, 10, 14, 2, 1.5, 6])
tempo = np.array([1, 2, 4, 6, 7, 10])
```

```python
dados = np.column_stack((corrente_eletrica, tempo))
```

```python
modelo = IsolationForest(contamination=0.2)
outliers = modelo.fit_predict(dados)
```

```python
indices_outliers = np.where(outliers == -1)[0]
```

```python
plt.scatter(tempo, corrente_eletrica, c='blue', label='Normal')
plt.scatter(tempo[indices_outliers], corrente_eletrica[indices_outliers], c='red', label='Anomalia')
plt.xlabel('Tempo')
plt.ylabel('Corrente Elétrica')
plt.legend()
plt.title('Detecção de Anomalias com Isolation Forest')
plt.show()
```
