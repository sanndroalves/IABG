```python
import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
leucocitos = np.array([2000, 4000, 5000, 6500])
plaquetas = np.array([100000, 20000, 80000, 145000])
linfocitos = np.array([2.3, 4.5, 6.5, 4.4])
```

```python
dados = np.column_stack((leucocitos, plaquetas, linfocitos))
```

```python
dbscan = DBSCAN(eps=1500, min_samples=2)  # Ajuste os parâmetros conforme necessário
grupos = dbscan.fit_predict(dados)
```

```python
print("Grupos de Pacientes:")
for i, grupo in enumerate(grupos):
    print(f"Paciente {i + 1}: Grupo {grupo}")
```

```python
plt.scatter(leucocitos, plaquetas, c=grupos, cmap='viridis', edgecolor='k')
plt.xlabel('Leucócitos')
plt.ylabel('Plaquetas')
plt.title('Agrupamento de Pacientes por Hemograma (Leucócitos vs. Plaquetas)')
plt.show()
```
