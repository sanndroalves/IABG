```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = {
    'Dia da Semana': [1, 2, 3, 4, 5, 6, 7],
    'Nº de Clientes': [15, 20, 30, 100, 350, 500, 700]
}

df = pd.DataFrame(data)
```

```python
X = df[['Nº de Clientes']].values

kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
print("DataFrame com Clusters:")
print(df)
```

```python
plt.scatter(df['Dia da Semana'], df['Nº de Clientes'], c=df['Cluster'], cmap='viridis', edgecolor='k')
plt.title('Agrupamento de Dias de Semana por Número de Clientes')
plt.xlabel('Dia da Semana')
plt.ylabel('Nº de Clientes')
plt.show()
```
