```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = pd.DataFrame({
    'Consumo de Energia (kWh)': np.array([1000,3000,1500,2500,1200,3500]),
    'Tempo de Operação (horas)': np.array([200,500,300,400,250, 600]),
    'Tipo de Energia': np.array(['Elétrica', 'Gás', 'Elétrica', 'Gás', 'Elétrica', 'Gás']),
    'Classificação': np.array(['Eficiente', 'Ineficiente', 'Eficiente', 'Ineficiente', 'Eficiente', 'Ineficiente'])
})
```

```python
tipo = np.array(['Elétrica', 'Gás', 'Elétrica', 'Gás', 'Elétrica', 'Gás'])
classificacao = np.array(['Eficiente', 'Ineficiente', 'Eficiente', 'Ineficiente', 'Eficiente', 'Ineficiente'])
```

```python
dados.head(6)
```

```python
tipo = np.array([1,0,1,0,1,0])
classificacao = np.array(['Eficiente', 'Ineficiente', 'Eficiente', 'Ineficiente', 'Eficiente', 'Ineficiente'])
```

```python
tipo_treino, tipo_teste, classificacao_treino, classificacao_teste = train_test_split(tipo.reshape(-1,1),classificacao.reshape(-1,1),test_size=0.3,random_state=42)                                                                      
```

```python
modelo = DecisionTreeClassifier()
modelo.fit(tipo_treino,classificacao_treino)
```

```python
previsoes = modelo.predict(tipo_teste)
```

```python
print(previsoes)
```

```python
fig = plt.figure(figsize=(10,8))
tree.plot_tree(modelo,feature_names= tipo.tolist(), class_names = classificacao.tolist(), filled= True)
```
