```python
anos = np.array([2, 6, 1, 4, 3, 5])
classificacao = np.array(['Necessita', 'Necessita', 'Necessita', 'Não Necessita', 'Necessita', 'Não Necessita'])
```

```python
anos_treino, anos_teste, classificacao_treino, classificacao_teste = train_test_split(anos.reshape(-1,1),classificacao.reshape(-1,1),test_size=0.3,random_state=42)                                                                      
```

```python
modelo = DecisionTreeClassifier()
modelo.fit(anos_treino,classificacao_treino)
```

```python
previsoes = modelo.predict(anos_teste)
```

```python
print(previsoes)
```

```python
fig = plt.figure(figsize=(10,8))
tree.plot_tree(modelo,feature_names= anos.tolist(), class_names = classificacao.tolist(), filled= True)
```
