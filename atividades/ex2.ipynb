```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
qtd = np.array([100, 200, 300, 400, 500, 600, 700]).reshape(-1, 1)
temp = np.array([30,35,40,45,60,100,150])

modelo = LinearRegression()
modelo.fit(qtd, temp)

qtd_teste = np.array([25,70,158,147,259,325,524]).reshape(-1, 1)
temp_prevista = modelo.predict(qtd_teste)
```

```python
for i in range(len(qtd_teste)):
    print(f"QTD: {qtd_teste[i][0]} - TEMP previsto: {temp_prevista[i]}º")
plt.scatter(qtd, temp, color="blue", label="Dados de Treinamento")
plt.plot(qtd_teste, temp_prevista, color = 'red', label = 'Dados de teste')
plt.xlabel("QUANTIDADE")
plt.ylabel("TEMPERATURA")
plt.title("Relação entre qtd e temperatura")
plt.legend()
plt.grid(True)
plt.show()
```
