```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
renda = np.array([1200, 5000, 8000, 10000, 30000]).reshape(-1, 1)
gastos = np.array([750, 1500, 4000, 6000, 18000])

modelo = LinearRegression()
modelo.fit(renda, gastos)

renda_teste = np.array([1400, 2541, 7581, 15420, 28965]).reshape(-1, 1)
gasto_previsto = modelo.predict(renda_teste)
```

```python
for i in range(len(renda_teste)):
    print(f"RENDA: {renda_teste[i][0]} - VALOR previsto: R${gasto_previsto[i]}")
plt.scatter(renda, gastos, color="blue", label="Dados de Treinamento")
plt.plot(renda_teste, gasto_previsto, color = 'red', label = 'Dados de teste')
plt.xlabel("RENDA (R$)")
plt.ylabel("GASTOS (R$)")
plt.title("Relação entre RENDA e GASTOS")
plt.legend()
plt.grid(True)
plt.show()
```
