```python
dados = pd.DataFrame({
    'Velocidade de Operação': np.array(['10', '5', '8', '6', '12', '4']),
    'Complexidade da Tarefa': np.array(['Baixa', 'Alta', 'Média', 'Alta', 'Baixa', 'Alta']),
    'Manutenção Necessária': np.array(['Baixa', 'Alta', 'Média', 'Alta', 'Baixa', 'Média']),
    'Classificação': np.array(['Montagem', 'Teste', 'Montagem', 'Teste', 'Montagem', 'Teste'])
})
```

```python
comple = np.array(['Baixa', 'Alta', 'Média', 'Alta', 'Baixa', 'Alta'])
classificacao = np.array(['Montagem', 'Teste', 'Montagem', 'Teste', 'Montagem', 'Teste'])
```

```python
prec = np.array([0,2,1,2,0,0])
classificacao = np.array(['Montagem', 'Teste', 'Montagem', 'Teste', 'Montagem', 'Teste'])
```

```python
precisao_treino, precisao_teste, classificacao_treino, classificacao_teste = train_test_split(prec.reshape(-1,1),classificacao.reshape(-1,1),test_size=0.3,random_state=42)                                                                      
```

```python
modelo = DecisionTreeClassifier()
modelo.fit(precisao_treino,classificacao_treino)
```

```python
previsoes = modelo.predict(precisao_teste)
```

```python
print(previsoes)
```

```python
fig = plt.figure(figsize=(10,8))
tree.plot_tree(modelo,feature_names= prec.tolist(), class_names = classificacao.tolist(), filled= True)
```
