```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dados_x = np.linspace(0,10,10)
dados_y = dados_x * dados_x
```

```python
print(dados_y)
```

```python
plt.figure(figsize=(10,5), facecolor = 'gray') 

Ax = plt.axes()
Ax.set_facecolor('white')

plt.plot(dados_x, dados_y,
         color = 'red',
         markersize = 5,
         marker = 'o',
         linestyle = 'dashdot',
         markeredgecolor = 'black',
         )
plt.xlabel("Eixo X", fontweight = 'bold',
           fontsize = 'large', fontfamily = 'Fantasy',
           color = 'blue')
plt.ylabel("Eixo Y", fontweight = 'bold',
           fontsize = 'large', fontfamily = 'Fantasy',
           color = 'blue')
plt.title("Grafico simples", fontweight = 'bold',
           fontsize = 'large', fontfamily = 'Fantasy',
           color = 'blue')
plt.legend("Numero aleatorios", 
           loc = "upper center", 
           bbox_to_anchor=(1,0,0.0,0.2), fontsize="large",
           shadow = True, borderpad = 1)
```

```python
plt.rcParams.update({'font.size':15})
plt.figure(figsize=(10,10))
linhas = 2
colunas = 2 
plt.subplot(linhas,colunas,1)
plt.plot(dados_x, "r--")
plt.title("Grafico 1")

plt.subplot(linhas,colunas,2)
plt.plot(dados_y, "b--")
plt.title("Grafico 2")

plt.subplot(linhas,colunas,3)
plt.plot(dados_y, "y--")
plt.title("Grafico 3")

plt.subplot(linhas,colunas,4)
plt.plot(dados_x, "g--")
plt.title("Grafico 4")


plt.savefig('Grafico simples.jpg')
```
