```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
vazao = np.array([10, 20, 30, 5, 35, 40]).reshape(-1, 1)
nivel = np.array([50, 60, 70, 45, 80, 85])

modelo = LinearRegression()
modelo.fit(vazao, nivel)

vazao_teste = np.array([1000,2000,500,3000,5000]).reshape(-1, 1)
nivel_previsao = modelo.predict(vazao_teste)
```

```python
for i in range(len(vazao_teste)):
    print(f"Vazao: {vazao_teste[i][0]} m3/min - Nível previsto: {nivel_previsao[i]} metros")
plt.scatter(vazao, nivel, color="blue", label="Dados de Treinamento")
plt.plot(vazao_teste, nivel_previsao, color = 'red', label = 'Dados de teste')
plt.xlabel("Vazao de Água (m3/min)")
plt.ylabel("Nível do tanque (metros)")
plt.title("Relação entre vazão de água e nível do tanque")
plt.legend()
plt.grid(True)
plt.show()


```
