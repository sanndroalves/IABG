```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
area = np.array([50,100,120,300]).reshape(-1, 1)
valor = np.array([180.000, 300.000, 375.000, 600.000])

modelo = LinearRegression()
modelo.fit(area, valor)

area_teste = np.array([47, 250, 365, 147]).reshape(-1, 1)
valor_previsto = modelo.predict(area_teste)
```

```python
for i in range(len(area_teste)):
    print(f"AREA: {area_teste[i][0]} - VALOR previsto: R${valor_previsto[i]}")
plt.scatter(area, valor, color="blue", label="Dados de Treinamento")
plt.plot(area_teste, valor_previsto, color = 'red', label = 'Dados de teste')
plt.xlabel("Area")
plt.ylabel("Valor (R$)")
plt.title("Relação entre area e valor")
plt.legend()
plt.grid(True)
plt.show()
```
