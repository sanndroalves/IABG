```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
meses = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
consumo = np.array([50,140,30,75,123])

modelo = LinearRegression()
modelo.fit(meses, consumo)

meses_teste = np.array([6,7,8,9,10]).reshape(-1, 1)
consumo_previsto = modelo.predict(meses_teste)
```

```python
for i in range(len(meses_teste)):
    print(f"MÊS: {meses_teste[i][0]} - CONSUMO previsto: R${consumo_previsto[i]}")
plt.scatter(meses, consumo, color="blue", label="Dados de Treinamento")
plt.plot(meses_teste, consumo_previsto, color = 'red', label = 'Dados de teste')
plt.xlabel("MES")
plt.ylabel("CONUSMO kMh")
plt.title("Relação entre mes e consumo")
plt.legend()
plt.grid(True)
plt.show()
```
