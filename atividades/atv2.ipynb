```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = pd.DataFrame({
    'Idade (anos)': np.array([5,10,3,8,1,15]),
    'Histórico de Manutenção': np.array(['Bom', 'Ruim', 'Excelente', 'Regular', 'Excelente', 'Ruim']),
    'Número de Falhas Anteriores': np.array([0,3,0,2,0,5]),
    'Nível de Automação': np.array(['Alto', 'Baixo', 'Médio', 'Alto', 'Médio', 'Baixo']),
    'Classificação': np.array(['Segura', 'Não Segura', 'Segura', 'Não Segura', 'Segura', 'Não Segura'])
})
```

```python
nivel = np.array(['Alto', 'Baixo', 'Médio', 'Alto', 'Médio', 'Baixo'])
classificacao = np.array(['Segura', 'Não Segura', 'Segura', 'Não Segura', 'Segura', 'Não Segura'])
```

```python
dados.head(6)
```

```python
nivel = np.array([2,0,1,2,1,0])
classificacao = np.array(['Segura', 'Não Segura', 'Segura', 'Não Segura', 'Segura', 'Não Segura'])
```

```python
nivel_treino, nivel_teste, classificacao_treino, classificacao_teste = train_test_split(nivel.reshape(-1,1),classificacao.reshape(-1,1),test_size=0.3,random_state=42)                                                                      
```

```python
modelo = DecisionTreeClassifier()
modelo.fit(nivel_treino,classificacao_treino)
```

```python
previsoes = modelo.predict(nivel_teste)
```

```python
print(previsoes)
```

```python
fig = plt.figure(figsize=(10,8))
tree.plot_tree(modelo,feature_names= nivel.tolist(), class_names = classificacao.tolist(), filled= True)
```
